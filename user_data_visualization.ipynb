```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load the CSV file into a pandas DataFrame
csv_file_path = 'user_data.csv'  # Replace with the correct file path if needed
df = pd.read_csv(csv_file_path)

# Display basic information about the dataset
print(df.info())
print(df.head())

# Perform some basic analysis, like summary statistics
print(df.describe())

# Check for any missing values
print(df.isnull().sum())

# Visualizations

# Distribution of total_income
plt.figure(figsize=(8, 6))
sns.histplot(df['total_income'], kde=True, bins=20, color='blue')
plt.title('Distribution of Total Income')
plt.xlabel('Total Income')
plt.ylabel('Frequency')
# Save the plot
plt.savefig('total_income_distribution.png')

# Expense distribution for different categories
expense_columns = ['food', 'transportation', 'utilities', 'entertainment', 'school_fees', 'shopping', 'healthcare', 'other']
expense_data = df[expense_columns].sum()

plt.figure(figsize=(10, 6))
expense_data.plot(kind='bar', color='orange')
plt.title('Total Expenses by Category')
plt.xlabel('Expense Categories')
plt.ylabel('Amount')
# Save the plot
plt.savefig('expense_by_category.png')

# C. Total income vs total expenses (scatter plot)
df['total_expenses'] = df[expense_columns].sum(axis=1)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['total_income'], y=df['total_expenses'], color='green')
plt.title('Total Income vs Total Expenses')
plt.xlabel('Total Income')
plt.ylabel('Total Expenses')

# Save the plot
plt.savefig('income_vs_expenses.png')

# Correlation heatmap of expenses
plt.figure(figsize=(10, 8))
sns.heatmap(df[expense_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Expenses')
# Save the plot
plt.savefig('expense_correlation_heatmap.png')

# Gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df, palette='Set2')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
# Save the plot
plt.savefig('gender_distribution.png')

# Display the final analysis and plots
plt.show()
```
